Sets (Operations):  

a. Create a set called my_set with following values:    
['a','e','i','o','u','a','a','i']
Analyse the output and provide explanation for the same.  

```python
my_set = ['a','e','i','o','u','a','a','i']
unique_my_set = set(my_set)
print(unique_my_set)
```

Explanation :  Set removes duplicate values so repeated 'a' and 'i' comes once at a time. set is unordered so values inside the set may interchange.

2. Find the performance category based on the input score using following criteria:
a. Above Average:  Score greater than 7
b. Average: Score between 4 and 7(both inclusive)
c. Below Average: Score lesser than 4

```python

score = int(input("Enter a score between 0-10:"))
if  score <0 or score>10:
       print("Enter a valid score between 0-10:")
elif score > 7 :
        print("Above average")
elif score >=4 and score <=7:
        print("Average")
else:
        print("Below average")



```
